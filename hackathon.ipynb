```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Dataset 1

```python
#Import Dataset 1.
df1 = pd.read_csv("1. annual-number-of-deaths-by-cause.csv")
```

##### Data Exploration and Cleaning.
- Extract African Countries.
- Delete columns with entirely missing data.
- Delete unnecessary columns in relation to our visualization.

```python
# Extract African Countries
df1 = df1[df1["Entity"] == "African Region (WHO)"] 
```

```python
#Dropping Columns with no data as well as unnecessary columns.
df1 = df1.drop(["Entity","Code", "Number of executions (Amnesty International)", "Terrorism (deaths)"], axis = 1)
df1.shape
```

```python
df1.info()
```

```python
df1.to_csv("df1.csv")
```

```python
df1.describe()
```

##### What  are the leading causes of death in Africa between 1990 and 2020?

```python
#Plotting the Diseases
plt.figure(figsize = (15,10))
plt.style.use("ggplot")
plt.plot(df1["Year"], df1["Cardiovascular diseases"], label = "Cardiovascular diseases")
plt.plot(df1["Year"], df1["Malaria"], label = "Malaria")
plt.plot(df1["Year"], df1["Lower respiratory infections"], label = "Lower respiratory infections")
plt.plot(df1["Year"], df1["Neonatal disorders"], label = "Neonatal disorders")
plt.plot(df1["Year"], df1["Diarrheal diseases"], label = "Diarrheal diseases")
plt.plot(df1["Year"], df1["Neoplasms"], label = "Neoplasms")
plt.title("Africa's Leading Causes of Death over the Years (1990 - 2020)")
plt.legend(loc='upper center')
plt.show()
```

##### Obervations
* Majority of the diseases are reducing in fatality for example Malaria, Lower Respiratory Infections and this is due a marked increase in treatment for those diseases.
* However diseases that build over time like Cardiovascular diseases, Neoplasms are on the rise due to lactancy by Africans to go for regular checkups.

##### NB
* The results in the graph do not visualize all the diseases instead the diseases woth the most fatalities and those whose numbers are increasing.

### Dataset 2

```python
#Importing the dataset
df2 = pd.read_csv("2. number-of-deaths-by-age-group.csv")
```

##### Data Exploration and Cleaning
- Extract African Countries.
- Delete columns with entirely missing data.
- Delete unnecessary columns in relation to our visualization.

```python
# Extract African Countries
df2 = df2[df2["Entity"] == "African Region (WHO)"] 
```

```python
#Dropping Columns with no data as well as unnecessary columns.
df2 = df2.drop(["Entity","Code"], axis = 1)
df2.shape
```

```python
df2.info()
```

```python
df2.describe()
```

```python
#Plotting Deaths by Age Groups
plt.figure(figsize = (15,10))
plt.style.use("ggplot")
plt.plot(df2["Year"], df2["Deaths 70+ years"], label = "Deaths 70+ years", color = "red")
plt.plot(df2["Year"], df2["Deaths 50-69 years"], label = "Deaths 50-69 years", color = "blue")
plt.plot(df2["Year"], df2["Deaths Age: 15-49 years"], label = "Deaths Age: 15-49 years", color = "black")
plt.plot(df2["Year"], df2["Deaths 5-14 years"], label = "Deaths 5-14 years", color = "green")
plt.plot(df2["Year"], df2["Deaths Under 5"], label = "Deaths Under 5", color = "purple")
plt.title("Deaths by Age Groups over the Years (1990 - 2020)")
plt.legend(loc='upper right')
plt.show() 
```

##### Obervations
* Deaths in the various age groups are on a downward spiral however age groups like the "Deaths 70+ years" and "Deaths 50-69 years" are registering a steady climb. This maybe attributed to susceptabiity to diseases like Cardiovascular diseases as identified in the earlier visualization.
* However deaths in "Deaths Under 5" and "Deaths 5-14 years" are slowly declining. This can be attributed to improving post natal care for children.
